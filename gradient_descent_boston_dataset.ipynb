```python
# lets add one feature for intercept "c" viz 1
df_new_1=df.copy()
df_new_1["c"]=1

print(df_new_1)
# lets make X and Y from this dataset df_new_1
X_new_1=df_new_1.values
# Y will remain same for this as we have in the boston dataset
print(X_new_1.shape)
```

```python
# lets train the data using the gradient descent funcction for multi-variables
def step_gradient_bostan(X,Y,learning_rate,m):
    features=X.shape[1]
    #print(features)
    m_slope=np.zeros(features)
    
    M=len(X)
    for i in range(M):
        for j in range(X_new_1.shape[1]):
            m_slope[j] += (-1/M)* (Y[i] - (m * X[i]).sum() )*X[i][j]
            
            
    new_m=m-learning_rate*m_slope
    
    return new_m
```

```python
def gd_multi_bostan(X,Y,learning_rate,num_of_itration):# this will calculate the m and c for run function after runing the n num of itration
    features=X.shape[1]
    
    m=np.zeros(features)

    
    for  i in range(num_of_itration):
        m=step_gradient_bostan(X,Y,learning_rate,m)
        print(i,"Cost",cost(X,Y,m))
    return m
```

```python
def cost(X,Y,m):
    total_cost=0
    M=len(X)
    for i in range(M):
        total_cost+= (1/(2*M))*((Y[i]-( sum(m*X[i])))**2)
    return total_cost
```

```python
# lets make a run function which will run the gradient descent function for n no of itrationa and having the learning rate and 
# return m and c:
def run_bostan(X,Y):
    
    learning_rate=0.05
    num_of_itration=300# if the no of itration will be more than a limit then it will be oversteping and you will get nan values
    # as if youn keep num_of_itration is 1000 then  it will give nan for run
    m=gd_multi_bostan(X,Y,learning_rate,num_of_itration)
    return (m)
```
